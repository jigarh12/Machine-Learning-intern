```python
normals = pd.Series(np.random.normal(size=10))
print(normals)
normals.plot()
```

```python
normals.cumsum().plot(grid=True)
```

```python
var = pd.DataFrame({'normal': np.random.normal(size=5), 
                       'gamma': np.random.gamma(1, size=5), 
                       'poisson': np.random.poisson(size=5)})
print(var)
var.cumsum(0).plot()
```

```python
var.cumsum().plot(subplots=True)
```

```python
var.cumsum(0).plot(secondary_y='normal')
```

```python
var.cumsum(0).plot(secondary_y='gamma')
```

```python
fig,axis=plt.subplots(nrows=3,ncols=1,figsize=(5,20))
```

```python
for i,v in enumerate(['normal','gamma','poisson']):
    var[v].cumsum(0).plot(ax=axis[i],title=v)
axis[0].set_ylabel('cumulative sum')
```

```python
fig,axis=plt.subplots(nrows=3,ncols=1,figsize=(4,16))
for i,v in enumerate(['normal','gamma','poisson']):
    var[v].cumsum(0).plot(ax=axis[i],title=v)
axis[0].set_ylabel('cumulative sum')
```

```python
fig,axis=plt.subplots(nrows=3,ncols=1,figsize=(4,16))
for i,v in enumerate(['normal','gamma','poisson']):
    var[v].cumsum(0).plot(ax=axis[i],title=v)
    axis[i].set_ylabel('cumulative sum')
    axis[i].set_xlabel('random number')
```
