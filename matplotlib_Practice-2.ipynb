```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
s = np.linspace(0, 20, 100)
print(s)
```

```python
plt.plot(s, np.cos(s))
plt.xlim(-1, 20)
plt.ylim(-1, 2);
```

```python
plt.plot(s, np.cos(s))
plt.xlim(-1, 20)
plt.ylim(-1, 1);
plt.title("Cosine Curve")
plt.xlabel("x")
plt.ylabel("cos(x)")
```

```python
x = np.random.normal(0, 0.8, 1000)
y = np.random.normal(-2, 1, 1000)
kwargs = dict(histtype='stepfilled', alpha=0.4, density=True, bins=10)
plt.hist(x, **kwargs)
plt.hist(y, **kwargs)
```

```python
x = np.random.normal(0, 2, 1000)
y = np.random.normal(-1, 2, 1000)
kwargs = dict(histtype='stepfilled', alpha=0.4, density=True, bins=100)
plt.hist(x, **kwargs)
plt.hist(y, **kwargs)
```
